```python
# ステップ関数
import numpy as np
import matplotlib.pylab as plt

def step_function(x):
    # x>0 ならTrueを返す。dttype = np.int boolをintに変換する。True-1,False-0
    return np.array(x > 0, dtype=int)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1) #y軸の範囲を指定
plt.show()
```

```python
# シグモイド関数
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1) #y軸の範囲を指定
plt.show()
```

```python
# 2関数の見比べ
y1 = sigmoid(x)
y2 = step_function(x)

# 同時にグラフ描画
plt.plot(x, y1, label="sigmoid")
plt.plot(x, y2, linestyle="--", label="step")
plt.ylim(-0.1, 1.1) # y軸の範囲を指定
plt.legend() # グラフの種類を表示(左上のやつ)
plt.show()
```

```python
# ReLU(Rectified Linear Unit)関数
def relu(x):
    return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.ylim(-1, 5) #y軸の範囲を指定
plt.show()
```
